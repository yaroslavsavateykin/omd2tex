```python
codeblock_lines = [
    "import omd2tex",
    "doc = Document().from_file('test.md')",
    "doc.to_latex_project()"
]

langs = ["python", "default","c++", "anykindoflanguage"]

for lan in langs:
    
    codeblock = CodeBlock(lan, codeblock_lines)
    print("Language used:", lan)
    print(codeblock.to_latex(), "\n"*2)
```
